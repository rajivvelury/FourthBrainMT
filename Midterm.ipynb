I.2 Data Pre-Processing and Cross-Validation

```python
#Now, we manually remove features from the dataframe that inspection would suggest are nonessential.  Based on correlations, it would make sense to trian on 'Administrative', 'Product_Related', 'Page Value', 'Bounce Rates', and 'Exit Rates,'  This is five features.
df_training = df.drop(['Month', 'VisitorType','Administrative_Duration', 'Informational', 'Informational_Duration', 'ProductRelated', 'ProductRelated_Duration', 'SpecialDay', 'OperatingSystems', 'Browser', 'Region', 'TrafficType', 'Weekend'], axis=1)
df_training.head()
```

```python
#Now, we correct for the class imbalance w.r.t. Revenue using SMOTE
```

```python
X = df_training.drop('Revenue', axis=1)
X.head()
y = df.Revenue
y.head()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

I.3 Modeling

```python
#Here, we perform and evaluate logistic regression

from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import confusion_matrix

clf = LogisticRegression(random_state=0).fit(X_train, y_train)
y_pred = clf.predict(X_test)
score = metrics.accuracy_score(y_test,y_pred)

confusion_matrix(y_test, y_pred)
print("Accuracy is " + str(100*score) +"%.")
print(confusion_matrix(y_test, y_pred))
```

```python
#Here, we use kNN and compare with the logistic regressio.
#Selection of k is important here.  We will loop over different ks and select based on highest accuracy.

from sklearn.neighbors import KNeighborsClassifier

scores = []
for k in range(1,50):
  knn = KNeighborsClassifier(n_neighbors = k)
  knn.fit(X_train,y_train)
  y_pred = knn.predict(X_test)
  score = metrics.accuracy_score(y_test, y_pred)
  scores.append(score)
max_score = max(scores)
max_k = scores.index(max_score)
print("Max k is " + str(max_k))
knn = KNeighborsClassifier(n_neighbors = max_k)
knn.fit(X_train,y_train)
y_pred = knn.predict(X_test)
score = metrics.accuracy_score(y_test, y_pred)
scores.append(score)
print("Accuracy is " + str(100*score) +"%.")
print(confusion_matrix(y_test, y_pred))
```

II. Customer Segmentation

```python
#Here, we perform k-means clustering with scikit-learn
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2, random_state=0).fit(X)
print(len(kmeans.cluster_centers_))
np.bincount(kmeans.labels_)

new_column = kmeans.predict(X)
data = df.copy()
data['Cluster'] = new_column
data.head()
data_cluster0 = data[data['Cluster']==0]
data_cluster1 = data[data['Cluster']!=0]
data_cluster0.describe()
```
